".isnull()" -> checks every cell and returns true if the cell is empty (missing), False if it has a value
".sum()" ->  counts how many true values are there / exists per column
".sum().sum()" -> The second sum() adds up all the missing values across all columns

```python
print("Missing Values in each column:")
print(df.isnull().sum())

print("\nTotal Missing Values:", df.isnull().sum().sum())
```

```python
cs_students = df[df['department']=='Computer Science']
print(f"Number Of Computer Science students: {len(cs_students)}")

print("\nComputer Science Students:")
print(cs_students[['name','math_score','programming_score','science_score','english_score','attendance_percentage']])
```

```python
high_attendance=df[df['attendance_percentage']>90]

print(f"Number of students with attendance above 90% : {len(high_attendance)}")

print(high_attendance[['name','department','attendance_percentage']])
```

```python
df['total_score'] = df['math_score'] + df['programming_score'] + df['science_score'] + df['english_score']
top_student = df.loc[df['total_score'].idxmax()]

print("Top Student:")
print(f"Name: {top_student['name']}")
print(f"Department: {top_student['department']}")
print(f"Total Score: {top_student['total_score']}")
print(f"Attendance: {top_student['attendance_percentage']}%")
```

```python
dept_avg = df.groupby('department')['total_score'].mean().round(2)

print("Average Scores Per Department:")
print(dept_avg)
```

```python
# In python there is no descending function so why use ascending=False.
df_sorted = df.sort_values(by='total_score', ascending=False)

print("Top 10 Students by Total Score:")
print(df_sorted[['name', 'department' , 'total_score' , 'attendance_percentage']].head(10))
```

```python
print("\nDATASET OVERVIEW")
print("-" * 30)

print(f"Total number of students : {df.shape[0]}")
print(f"Total number of columns  : {df.shape[1]}")
print(f"Unique departments       : {df['department'].nunique()}")
```

```python
print("\nDEPARTMENT-WISE ANALYSIS")
print("-" * 30)

dept_count = df['department'].value_counts()

for dept, count in dept_count.items():
    print(f"{dept} : {count} students")
```

```python
print("\nSCORE ANALYSIS")
print("-" * 30)

print(f"Highest Math Score            : {df['math_score'].max()}")
print(f"Lowest Math Score             : {df['math_score'].min()}")
print(f"Average Programming Score     : {df['programming_score'].mean():.2f}")
print(f"Average Attendance Percentage : {df['attendance_percentage'].mean():.2f}")
```

```python
print("\nGENDER-WISE ANALYSIS")
print("-" * 30)

gender_avg = df.groupby('gender')['total_score'].mean().round(2)

for gender, avg in gender_avg.items():
    print(f"{gender} : {avg}")
```

```python
print("\nTOP 5 STUDENTS BASED ON TOTAL SCORE")
print("-" * 50)

top_students = df.sort_values(by='total_score', ascending=False).head(5)

for index, row in top_students.iterrows():
    print(f"Name : {row['name']}")
    print(f"Department : {row['department']}")
    print(f"Total Score : {row['total_score']}")
    print("-" * 30)
```

```python
print("\nSTUDENTS WITH ATTENDANCE BELOW 75%")
print("-" * 40)

low_attendance = df[df['attendance_percentage'] < 75]

if len(low_attendance) > 0:

    for index, row in low_attendance.iterrows():
        print(f"Name : {row['name']}")
        print(f"Department : {row['department']}")
        print(f"Attendance Percentage : {row['attendance_percentage']}%")
        print("-" * 30)

else:
    print("All students have attendance above 75%")
```

```python
#
# 1. PROJECT TITLE
#

print("=" * 50)
print("      STUDENT PERFORMANCE EXPLORER")
print("=" * 50)

#
# 2. DATASET OVERVIEW
#

print("\nDATASET OVERVIEW")
print("-" * 30)

print(f"Total number of students : {df.shape[0]}")
print(f"Total number of columns  : {df.shape[1]}")
print(f"Unique departments       : {df['department'].nunique()}")

#
# 3. DEPARTMENT-WISE ANALYSIS
#

print("\nDEPARTMENT-WISE ANALYSIS")
print("-" * 30)

dept_count = df['department'].value_counts()

for dept, count in dept_count.items():
    print(f"{dept} : {count} students")

#
# 4. SCORE ANALYSIS
#

print("\nSCORE ANALYSIS")
print("-" * 30)

print(f"Highest Math Score            : {df['math_score'].max()}")
print(f"Lowest Math Score             : {df['math_score'].min()}")
print(f"Average Programming Score     : {df['programming_score'].mean():.2f}")
print(f"Average Attendance Percentage : {df['attendance_percentage'].mean():.2f}")

#
# 5. GENDER-WISE ANALYSIS
#

print("\nGENDER-WISE ANALYSIS")
print("-" * 30)

gender_avg = df.groupby('gender')['total_score'].mean().round(2)

for gender, avg in gender_avg.items():
    print(f"{gender} : {avg}")

#
# 6. TOP 5 STUDENTS
#

print("\nTOP 5 STUDENTS BASED ON TOTAL SCORE")
print("-" * 50)

top_students = df.sort_values(by='total_score', ascending=False).head(5)

for index, row in top_students.iterrows():
    print(f"Name : {row['name']}")
    print(f"Department : {row['department']}")
    print(f"Total Score : {row['total_score']}")
    print("-" * 30)

#
# 7. ATTENDANCE ANALYSIS
#

print("\nSTUDENTS WITH ATTENDANCE BELOW 75%")
print("-" * 40)

low_attendance = df[df['attendance_percentage'] < 75]

if len(low_attendance) > 0:

    for index, row in low_attendance.iterrows():
        print(f"Name : {row['name']}")
        print(f"Department : {row['department']}")
        print(f"Attendance Percentage : {row['attendance_percentage']}%")
        print("-" * 30)

else:
    print("All students have attendance above 75%")

```
